```python
import plotly.graph_objects as go

# usamos un df de plotly con precios de stocks 
df_stocks = px.data.stocks()

# graficamos una lnea de tendencia del precio de las acciones de google a lo largo del tiempo
px.line(df_stocks, x='date', y='GOOG', labels={'x': 'Date', 'y': 'Price'})
```

```python
# Graficamos el precio de las acciones de Google vs Apple 
px.line(df_stocks, x='date', y= ['GOOG', 'AAPL'], 
        labels={'x':'Date', 'y':'Price'}, 
        title='Apple vs. Google')
```

```python
# Graficamos el precio de las acciones de Google, Apple y Amazon

fig = go.Figure()
fig.add_trace(go.Scatter(x=df_stocks.date, y=df_stocks.AAPL,
                        mode='lines', name='Apple'))

fig.add_trace(go.Scatter(x=df_stocks.date, y=df_stocks.AMZN,
                        mode='lines+markers', name='Amazon'))


fig.add_trace(go.Scatter(x=df_stocks.date, y=df_stocks.GOOG,
                        mode='lines+markers', name='Google',
                        line=dict(color='firebrick', width=2, dash='dashdot')))

fig.update_layout(title='Stock Price Data 2018-2020',
                  xaxis_title='Price', yaxis_title='Date')
```

```python
# Graficamos el precio de las acciones de Google, Apple y Amazon

fig = go.Figure()
fig.add_trace(go.Scatter(x=df_stocks.date, y=df_stocks.AAPL,
                        mode='lines', name='Apple'))

fig.add_trace(go.Scatter(x=df_stocks.date, y=df_stocks.AMZN,
                        mode='lines+markers', name='Amazon'))


fig.add_trace(go.Scatter(x=df_stocks.date, y=df_stocks.GOOG,
                        mode='lines+markers', name='Google',
                        line=dict(color='firebrick', width=2, dash='dashdot')))

fig.update_layout(title='Stock Price Data 2018-2020',
                  xaxis_title='Price', yaxis_title='Date')

# Practicamos personalizacin del grfico
fig.update_layout(
    xaxis=dict(
        showline=True, 
        showgrid=False, 
        showticklabels=True, 
        linecolor='rgb(204,204,204)',
        linewidth=2, 
        ticks='outside', 
        tickfont=dict(
            family='Arial', 
            size=12, 
            color='rgb(82,82,82)'
        )
    ),
    yaxis=dict(
        showgrid=False, 
        zeroline=False, 
        showline=False,
        showticklabels=False
    ),
    autosize=False, 
    margin=dict(
        autoexpand=False, 
        l=100, 
        r=20, 
        t=110
    ),
    showlegend=False, 
    plot_bgcolor='white'
)
```
