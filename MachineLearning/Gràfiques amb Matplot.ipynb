```python
import pandas as pd
insurance_data = pd.read_csv('insurance.csv')
insurance_data.head(10)
```

```python
# bmi √©s el valor de Body Mass Index i charges √©s l‚Äôimport de risc de l‚Äôasseguran√ßa.
plt.scatter(insurance_data['bmi'], insurance_data['charges'],)
plt.xlabel('bmi')
plt.ylabel('charges')
plt.show()
```

```python
import seaborn as sns

sns.scatterplot(x=insurance_data['bmi'],y=insurance_data['charges'])
```

```python
# esbrinem si hi ha una correlaci√≥ entre els valors bmi i el preu de la p√≤lissa

sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])

""" Segons el gr√†fic, efectivament hi ha una correlaci√≥ positiva. A mesura que el valor bmi d‚Äôuna p
ersona augmenta, tamb√© creix el preu. """
```

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])

"""Aqu√≠ veiem que, clarament, si la persona √©s fumadora, el preu de la seva asseguran√ßa ser√† m√©s alt. """
```

```python
# volem veure si la relaci√≥ entre bmi i charges √©s diferent, en funci√≥ de si una persona √©s fumadora o no

sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)

"""Efectivament, veiem que, per a una persona fumadora, el preu de la seva p√≤lissa augmenta considerablement quan t√© un bmi m√©s alt, 
mentre que les persones no fumadores tenen un augment molt m√©s petit. """
```

```python
# buscamos relaci√≥n entre el precio de la p√≥liza y la condici√≥n de fumador
sns.swarmplot(x=insurance_data['smoker'], y=insurance_data['charges'])

""" Aquest gr√†fic ens ensenya que la gran majoria de persones no fumadores tenen un preu inferior al de les persones fumadores. """
```

```python
# indagamos en la relaci√≥n entre el monto de la p√≥liza y la regi√≥n

sns.swarmplot(x=insurance_data['region'], y=insurance_data['charges'])

""" En principi, la majoria de la poblaci√≥ no presenta diferents preus, segons la seva regi√≥. """
```

```python
# qu√® passa, si afegim a l‚Äôan√†lisi la variable smoke?

sns.swarmplot(x=insurance_data['region'],y=insurance_data['charges'], hue=insurance_data['smoker'])

"""Aqu√≠, una altra vegada, comprovem que el factor m√©s determinant per veure com varia el preu d‚Äôuna asseguran√ßa √©s saber si la persona √©s fumadora o no.""" 
```
