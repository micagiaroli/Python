```python
import pandas as pd

notes = pd.DataFrame([[6.5,6.8,5.8,7.9,8.8,'Barcelona'], [6.8,7.0,6.5,8.9,9.2,'Girona'],
                                                [7.8,6.2,6.5,4.1,8.0,'Tarragona'], [6.9,7.2,4.7,8.8,6.2,'Lleida']],
                                                columns=['Història','Matemàtiques','Física','Anglès','Biologia',
                                               'Ciutat'],index=['Joan','Laura','Enric','Georgina',])
notes
```

```python
print(notes['Física'][1])
print(notes['Ciutat'][0])
```

```python
# qui va aprobar física?

notes[notes.Física > 5]
```

```python
# canviem la nota de la Georgina perqué sí que va aprobar

notes['Física'][3] = 7.0
notes[notes.Física > 5]
```

```python
# seleccionem taula reduïda amb les notes de biologia

notes.loc[['Laura','Georgina'],['Biologia']]
```

```python
# afegim una columna amb les notes de cada estudiant i una altra amb la seva desviació

mitjanes = []
desviacio = []

for i in range(4):
    mitjana = notes.iloc[i, 0:5].mean()
    desv = notes.iloc[i, 0:5].std()
    mitjanes.append(mitjana)
    desviacio.append(desv)

notes['Mitjanes'] = mitjanes
notes['Desviació'] = desviacio

notes
```
