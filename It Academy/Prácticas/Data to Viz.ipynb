# Histograma con Seaborn: Displot()
Muestra la distribución de una variable numérica.

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')

sns.displot( data= df["sepal_length"], kde=True)
plt.title("Histograma: Longitud del sépalo")
```

# Histograma con Matplotlib hist()

```python
# library & dataset
import matplotlib.pyplot as plt
hours = [17, 20, 22, 25, 26, 27, 30, 31, 32, 38, 40, 40, 45, 55]

# Initialize layout
fig, ax = plt.subplots(figsize = (7, 7))

#plot
ax.hist(hours, bins=5, edgecolor="black");

plt.title("Histograma: Distribución de horas")
plt.xlabel("Hours")
plt.ylabel("Frequency")
```

# Gráfico de densidad para una variable numérica contínua

```python
# library & dataset
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('iris')

# Make default density plot
sns.kdeplot(df['sepal_width'])
plt.title("Gráfico de densidad")
```
