```python
import pandas as pd
import sqlite3

# leemos archivo csv y lo cargamos en un dataframe llamado data
data = pd.read_csv('C:/Users/mica_/OneDrive/Escritorio/Cursos/Anàlisi de dades BA/Prácticas/IMB SQL Data Science/menu.csv')

# establecemos conexión a una base de datos SQLite que sino existe, la creamos en el mismo directorio donde estamos ejecutando el script
conn = sqlite3.connect('Mcdonalds.db')

# guardamos el df "data" en una tabla de SQLite conectada a través de conn, llamamos la tabla "MCDONALDS_NUTRITION"
data.to_sql('MCDONALDS_NUTRITION', conn, if_exists='replace', index=False)
```

```python
df = pd.read_sql("SELECT * FROM MCDONALDS_NUTRITION", conn)
print(df)
```

## Which food item has maximum sodium content?

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
plot = sns.stripplot(x="Category", y="Sodium", data=df, hue='Category')
plt.setp(plot.get_xticklabels(), rotation=70)
plt.title('Sodium Content')
plt.show()
```
