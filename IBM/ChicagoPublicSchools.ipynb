### ¿Cuál es el puntaje más alto de seguridad?

```python
%sql select MAX(Safety_Score) AS MAX_SAFETY_SCORE from chicago_schools
```

### ¿Qué escuelas tienen ese mayor puntaje?

```python
%sql SELECT name_of_school, safety_score FROM chicago_schools WHERE safety_score = 99.0
```

### ¿Cuál es el top 10 de escuelas con el mayor "Average Student Attendance"?

```python
%%sql 
SELECT name_of_school, Average_Student_Attendance
FROM chicago_schools
ORDER BY Average_Student_Attendance DESC
LIMIT 10
```

### Listado de las 5 escuelas con el menor Average Student Attendance ordenado ascendente según la asistencia

```python
%%sql 
SELECT name_of_school, Average_Student_Attendance
FROM chicago_schools
ORDER BY Average_Student_Attendance 
NULLS LAST
LIMIT 5
```

### Remueve el signo de porcentaje de los resultados de la consulta anterior de la columna AVERAGE_STUDENT_ATTENDANCE

```python
%sql SELECT Name_of_School, REPLACE(Average_Student_Attendance, '%', '') \
     from CHICAGO_SCHOOLS \
     order by Average_Student_Attendance \
     NULLS LAST \
     LIMIT 5
```

###

### ¿Qué escuelas tienen el Average Student Attendance por debajo del 70%?

```python
%%sql
SELECT name_of_school, average_student_attendance
FROM chicago_schools
WHERE average_student_attendance < 70
```

### Obtiene la inscripción universitaria total para cada área comunitaria

```python
%%sql
SELECT community_area_name, SUM(college_enrollment) AS total_enrollment
FROM chicago_schools
GROUP BY community_area_name
```

### Obtenga las 5 áreas comunitarias con la menor inscripción universitaria total ordenadas en orden ascendente

```python
%%sql
SELECT community_area_name, SUM(college_enrollment) AS total_enrollment
FROM chicago_schools
GROUP BY community_area_name
ORDER BY SUM(college_enrollment)
LIMIT 5
```

### Listado de las 5 escuelas con el menor índice de seguridad

```python
%%sql
SELECT name_of_school, safety_score
FROM chicago_schools
ORDER BY safety_score
NULLS LAST
LIMIT 5
```

### Obtenga el índice de dificultades para el área comunitaria de la escuela que tiene una matrícula universitaria de 4368

```python
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DB0201EN-SkillsNetwork/labs/FinalModule_Coursera_V5/data/ChicagoCensusData.csv")
df.to_sql("CENSUS_DATA", con, if_exists='replace', index=False,method="multi")
```

```python
%%sql 
select hardship_index from CENSUS_DATA CD, CHICAGO_SCHOOLS CS 
where CD.community_area_number = CS.community_area_number 
and college_enrollment = 4368
```
